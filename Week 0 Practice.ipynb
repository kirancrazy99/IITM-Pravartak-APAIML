## 1) This Python script introduces fundamental data handling, statistical summary, and basic visualization using a small example dataset. The dataset includes fictional information about four individuals—their names, ages, and test scores.

In this example, you'll:

1. Create a pandas DataFrame – to organize and display tabular data.
2. Print and summarize the data – using built-in pandas functions.
3. Visualize data with a scatter plot – using matplotlib.
4. Fit a simple linear regression – with scikit-learn, and plot the regression line.

This is a great starting point for beginners looking to get comfortable with Python’s data science stack

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
import numpy as np
```

```python
# Create sample data
data = pd.DataFrame({
    'Name': ['Alice', 'Bob', 'Charlie', 'David'],
    'Age': [25, 30, 22, 35],
    'Score': [85, 90, 88, 95]
})
```

```python
# Display the data
print("Data Frame:")
print(data)
```

```python
# Summary Statistics

print('\nSummary Statistics')
data.describe()
```

```python
X = data[['Age']]
Y = data['Score']

model = LinearRegression()
model.fit(X,Y)
```

```python
x_vals = np.linspace(data['Age'].min(),data['Age'].max(),100).reshape(-1,1)
x_vals_df = pd.DataFrame(x_vals,columns=['Age'])
y_pred = model.predict(x_vals_df)
y_pred
```

```python
plt.figure(figsize=(8,5))
plt.scatter(data['Age'],data['Score'],color='red',label='Data Points')
plt.plot(x_vals_df,y_pred,color='green',label='Regression LIne')
plt.title('Score vs Age')
plt.xlabel('Age')
plt.ylabel('Score')
plt.grid()
plt.legend()
plt.show()
```
