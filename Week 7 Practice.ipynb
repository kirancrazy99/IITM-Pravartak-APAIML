```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import style

style.use('ggplot')
titanic = sns.load_dataset('titanic')

titanic.shape
```

```python
titanic.head()
```

```python
titanic.info()
```

```python
titanic.dropna(subset=['age'],inplace=True)
```

```python
sns.histplot(data=titanic,x='age')
plt.title('Age distribution')
plt.show()
```

```python
sns.countplot(data=titanic,x='sex')
plt.title('Cout of passengers by gender')
plt.show()
```

## Bivariate Analysis

```python
sns.scatterplot(data=titanic, x='age',y='fare',hue='survived')
plt.title('Age vs Fare (Colored by survival)')
plt.show()
```

```python
sns.barplot(data=titanic,x='class',y='fare',errorbar=None)
plt.title('Average Fare by class')
plt.show()
```

```python
sns.boxplot(data=titanic,x='class',y='age',hue='sex')
plt.show()
```

## Multivariate Analysis

```python
corr = titanic.corr(numeric_only=True)
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

```python
sns.pairplot(data=titanic,vars=['age','fare','survived'],hue='class')
```

```python
# Changing the style and palette
sns.set_theme(style="whitegrid", palette="pastel")

# Customized count plot
sns.countplot(data=titanic, x="embark_town", hue="class")
plt.title("Passenger Count by Embarkation Town and Class")
plt.xlabel("Embarkation Town")
plt.ylabel("Count")
plt.legend(title="Class")
plt.show()
```

## # Faceted Visualizations

```python
g = sns.FacetGrid(titanic, row="sex", col="class", hue="survived", margin_titles=True)
g.map(sns.scatterplot, "age", "fare", alpha=0.7)
g.add_legend()
plt.show()
```

```python
# Relational plot with line segments for fare by age
sns.relplot(
    data=titanic, x="age", y="fare", kind="line",
    hue="class", style="sex", markers=True, dashes=False
)
plt.title("Fare by Age for Different Classes")
plt.show()
```

```python
# FacetGrid with segments for survival and fare
g = sns.FacetGrid(titanic, col="sex", row="class", hue="survived", height=4, aspect=1.5)
g.map(sns.lineplot, "age", "fare", alpha=0.7)
g.add_legend()

plt.show()
```

```python
import warnings
warnings.filterwarnings('ignore')
# Strip plot with jitter
plt.figure(figsize=(14,6))
sns.stripplot(data=titanic, x="class", y="age", jitter=True, hue="sex", dodge=True)
plt.title("Age Distribution by Class and Gender")
plt.show()

# Swarm plot
plt.figure(figsize=(14,6))
sns.swarmplot(data=titanic, x="class", y="fare", hue="sex", dodge=True)
plt.title("Fare Distribution by Class and Gender")
plt.show()
```

```python
import inspect
def greet(name: str, age: int = 30) -> str:
    """Greet a person by name and age."""
    return f"Hello, {name}! You are {age} years old."
signature = inspect.signature(greet)
print(signature) # Output: (name: str, age: int = 30) ->
str
for param in signature.parameters.values():
    print(f"Parameter: {param.name}, Type:{param.annotation}, Default: {param.default}")
```
